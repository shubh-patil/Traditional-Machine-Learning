```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cp = pd.read_csv('CarPrice.csv')
cp.head()
```

```python
#Data Preprocessing
cp.info()
```

```python
#check for NA values
cp.isnull().sum()

#no missing values
```

```python
cp.drop('car_ID', axis = 1, inplace= True)
```

```python
cp.head()
```

```python
#exploring each feature
cp['fuelsystem'].value_counts()

#getting list of columns that are not int or float excluding price

Str_cols = [features for features in cp.columns if cp[features].dtypes=='str']

print(Str_cols)

Str_cols.remove('CarName')
```

```python
#encoding all features

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop = 'first', sparse_output= False)
encoded_array = encoder.fit_transform(cp[Str_cols])

encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(Str_cols))



encoded_df.head()
```

```python
#dropping categorical features
cp.drop(Str_cols, axis = 1, inplace= True)
```

```python
#drop car name as well
cp.drop('CarName', axis = 1, inplace= True)
```

```python
#concate both int data and encoded data
cp = pd.concat([cp, encoded_df], axis= 1)
cp.info()
```
